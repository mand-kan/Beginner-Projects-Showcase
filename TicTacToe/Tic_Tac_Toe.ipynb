```
move=['O','X']
position = 0
replay=True

import random

def player_name():
  global playername
  playername={'A':'','B':''}
  playername['A']=input("Hi Player A, What's your name?")
  playername['B']=input("Hi Player B, What's your name?")

def reset():
  global myboard
  global playermark
  global full 
  global win
  global playing
  global replay_input

  myboard=[]
  playermark={'A':'','B':''}
  full = False
  win = False
  playing =''
  replay_input = ''


def player_input():
  global playermark
  global move
  while playermark['A'] not in move:
    playermark['A']=input(f"Hi {playername['A']}, Do you want O or X? ").upper()

  if playermark['A'] == move[0]:
      playermark['B'] = move[1]
  else:
      playermark['B']=move[0]

  print(f" {playername['A']} is {playermark['A']} and {playername['B']} is {playermark['B']}.")

#check int
def player_play():
  global position
  while True:
    position = (input(f'{playername[playing]}, Please enter a number(1-9):'))
    try:
      position = int(position)
      if position in range(1,10):
        break
      else:
        continue
    except ValueError:
      print("It's not a number")


def display_board(board):
    print('   |   |')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('   |   |')
    

def place_marker(board, mark, position):
    board[position]=mark
    display_board(board)


def random_play():
  global playing
  playing = random.choice(list(playermark))
  print(f"{playername[playing]} will start first.")


def check_full():
  global full
  if " " not in myboard:
    print("It's a draw!")
    full = True
    


def check_win(myboard,mark):
  global win
  global playing
  #check row
  if myboard[1]==myboard[2]==myboard[3]==mark:
    win = True
  elif myboard[3]==myboard[5]==myboard[6]==mark:
    win = True
  elif myboard[7]==myboard[8]==myboard[9]==mark:
    win = True
  #check column
  elif myboard[1]==myboard[4]==myboard[7]==mark:
    win = True
  elif myboard[2]==myboard[5]==myboard[8]==mark:
    win = True
  elif myboard[3]==myboard[6]==myboard[9]==mark:
    win = True
  #check diagonal
  elif myboard[1]==myboard[5]==myboard[9]==mark:
    win = True
  elif myboard[3]==myboard[5]==myboard[7]==mark:
    win = True
  
  if win==True:
    print(f"Congratulations! {playername[playing]} You win!")

def change_player():
    global playing 
    if playing=='A':
      playing='B'
    else:
      playing='A'

def check_replay():
  global replay
  if full == True or win==True:
    replay_input = input("Play Again? (Y/N)")
  while replay_input not in ['Y','N']:
    replay_input = input("Sorry I don't understand. Play Again? (Y/N)").upper()
  if replay_input =='Y':
    replay = True
  elif replay_input == 'N':
    replay = False
    print('Goodbye!')


#start
print("Welcome to Tic Tac Toe")
player_name()
while replay == True:
  reset()
  player_input()
  random_play()
  #set up
  for x in range(0,10):
    if x ==0:
     myboard.append('#')
    else: 
     myboard.append(' ')

  while full == False and win == False :
      player_play()
      if myboard[position] not in move:   
        place_marker(myboard, playermark[playing], position)
        check_full()
        check_win(myboard,playermark[playing])
        change_player()
      else:
        print("This is occupied!")
  check_replay()
  continue 
```
